```
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

from collections import Counter
```

```
plt.rcParams["figure.figsize"] = (10,15)
```

```
def prob_coin_toss(n_times):
  p_coin = {}
  toss = Counter(np.random.randint(2, size=n_times))
  p_coin[0] = toss[0]/n_times
  p_coin[1] = toss[1]/n_times
  return p_coin
```

```
def sample_probabilities(n_times):
  return [prob_coin_toss(n_times) for n_times in range(1, n_times+1)]
```

```
def output_count(sample_prob):
  all_0 = []
  all_1 = []
  for i in sample_prob:
    all_0.append(i[0])
    all_1.append(i[1])
  return all_0, all_1
```

```
def output_head(n_times):
  sample_prob = sample_probabilities(n_times)
  output_0, output_1 = output_count(sample_prob)
  output_1.sort()

  mean_1 = np.mean(output_1)
  std_1 = np.std(output_1)

  return output_1, stats.norm.pdf(output_1, mean_1, std_1)
```

```
def output_tail(n_times):
  sample_prob = sample_probabilities(n_times)
  output_0, output_1 = output_count(sample_prob)
  output_0.sort()

  mean_0 = np.mean(output_0)
  std_0 = np.std(output_0)

  return output_0, stats.norm.pdf(output_0, mean_0, std_0)
```

**Comparision of normal distribution plot taking different values of samples**

```
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1)
fig.suptitle('Distribution of different samples of coin toss (head)')
o_1, pdf_1 = output_head(10)
ax1.set_title("N_samples=10")
ax1.plot(o_1, pdf_1)  

o_2, pdf_2 = output_head(100)
ax2.set_title("N_samples=100")
ax2.plot(o_2, pdf_2)

o_3, pdf_3 = output_head(1000)
ax3.set_title("N_samples=1000")
ax3.plot(o_3, pdf_3)

o_4, pdf_4 = output_head(10000)
ax4.set_title("N_samples=10000")
ax4.plot(o_4, pdf_4)
```

```
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1)
fig.suptitle('Distribution of different samples of coin toss (tail)')
o_1, pdf_1 = output_tail(10)
ax1.set_title("N_samples=10")
ax1.plot(o_1, pdf_1)  

o_2, pdf_2 = output_tail(100)
ax2.set_title("N_samples=100")
ax2.plot(o_2, pdf_2)

o_3, pdf_3 = output_tail(1000)
ax3.set_title("N_samples=1000")
ax3.plot(o_3, pdf_3)

o_4, pdf_4 = output_tail(10000)
ax4.set_title("N_samples=10000")
ax4.plot(o_4, pdf_4)
```
